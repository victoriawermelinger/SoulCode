```python
import pandas as pd
import numpy as np
from IPython.display import display

# Ler dados de um arquivo.csv
desgaste_funcionario = pd.read_csv('C:/Users/victoria/Documents/MeusProjetos- GIT/SoulCode/Bootcamp - ANALISE DE DADOS/Base de dados - Dados sobre desgaste de funcionarios.csv')
```

```python
#limpeza dos valores não nulos.
data_f = desgaste_funcionario.dropna(axis=1, thresh=2)
print(data_f)

#Visualizar os dados, de forma mais clara. 
display(desgaste_funcionario)
```

```python
#1. Frequencia de viagens
quantidade_viagem = desgaste_funcionario['Viagem de negócios'].value_counts()
display(quantidade_viagem)
```

```python
#2. Área de formação dos Funcionarios 
area_func = desgaste_funcionario['Área de Formação'].value_counts()
display(area_func)
```

```python
#3. Agrupando por 'Área de Formação' e 'Viagem de negócios' e contando o número de ocorrências

contagem_area_viagem = desgaste_funcionario.groupby(['Área de Formação','Viagem de negócios']).size().unstack()
display(contagem_area_viagem)
```

```python
#4. Departamento que tem mais viagens 
contagem_area_viagem = desgaste_funcionario.groupby(['Departamento','Viagem de negócios']).size().unstack()
display(contagem_area_viagem)
```

```python
#5. Quadro de formação por departamento 
contagem_dep_area = desgaste_funcionario.groupby(['Área de Formação', 'Departamento']).size().unstack()
display(contagem_dep_area)
```

```python
#6. Remuneração por ações da empresa para os funcionarios
def remuneração(row):
        if row['opção de remaunração variável em ações'] == 3:
            return 'Participação nas ações'
        elif row['opção de remaunração variável em ações'] == 2:
            return 'Recebe dividendos, mas não tem ação'
        elif row['opção de remaunração variável em ações'] == 1:
            return 'Recebe dividendos'
        else:
            return 'Não recebe'

desgaste_funcionario['Grau Remuneração'] = desgaste_funcionario.apply(remuneração, axis=1)
display(desgaste_funcionario[['Departamento', 'Área de Formação', 'Grau Remuneração']])
```

```python
#7. contagem da remuneração dos funcionários 
grau_remuneração = desgaste_funcionario[('Grau Remuneração')].value_counts()
display(grau_remuneração)
```

```python
#8. grau de satisfação dos funcionarios (Nível de Satisfação com o trabalho)
def Satisfação(row):
        if row['Satisfação com o trabalho'] == 4:
            return 'Muito Alta'
        elif row['Satisfação com o trabalho'] == 3:
            return 'Alta'
        elif row['Satisfação com o trabalho'] == 2:
            return 'Média'
        else:
            return 'Baixa'

desgaste_funcionario['Grau Satisfação'] = desgaste_funcionario.apply(Satisfação, axis=1)
display(desgaste_funcionario[['Departamento', 'Área de Formação', 'Grau Satisfação']])
```

```python
#9. contagem de Grau Satisfação dos funcionários 
grau_satisfação = desgaste_funcionario[('Grau Satisfação')].value_counts()
display(grau_satisfação)
```

```python
#10.  grau de escolaridade dos funcionarios 
def Escolaridade(row):
        if row['Escolaridade'] == 5:
            return 'Doutor'
        elif row['Escolaridade'] == 4:
            return 'Mestre'
        elif row['Escolaridade'] == 3:
            return 'Bacharelado'
        elif row['Escolaridade'] == 2:
            return 'Faculdade'
        else:
            return 'Abaixo da faculdade'

desgaste_funcionario['Grau Escolaridade'] = desgaste_funcionario.apply(Escolaridade, axis=1)

display(desgaste_funcionario[['Área de Formação', 'Grau Escolaridade']])
```

```python
#11. contagem de Grau escolaridade dos funcionários 
grau_escolaridade_func = desgaste_funcionario['Grau Escolaridade'].value_counts()
display(grau_escolaridade_func)
```

```python
#12. grau de Experiência dos funcionarios 

def experiência(row):
    if row['Total de anos trabalhados'] > 0 and row['Total de anos trabalhados'] <= 5:
        return 'JV/Estágiario'
    elif row['Total de anos trabalhados'] > 5 and row['Total de anos trabalhados'] <= 10:
        return 'Junior'
    elif row['Total de anos trabalhados'] > 10 and row['Total de anos trabalhados'] <= 20:
        return 'Pleno'
    elif row['Total de anos trabalhados'] > 20:
        return 'Senior'
    else:
        return 'Sem experiência'

desgaste_funcionario['Grau experiência'] = desgaste_funcionario.apply(experiência, axis=1)

display(desgaste_funcionario[['Total de anos trabalhados', 'Área de Formação', 'Grau experiência']])
```

```python
#13. Contagem dos diferentes valores na coluna 'Grau experiência'
grau_experiência_func = desgaste_funcionario['Grau experiência'].value_counts()
display(grau_experiência_func)
```

```python
#14. Calcular a média dos anos trabalhados para cada grau de experiência
media_exp = desgaste_funcionario.groupby('Grau experiência')['Total de anos trabalhados'].mean()
display(media_exp)
```

```python
#15. Contar o número de funcionários em cada combinação de 'Grau experiência' e 'Departamento'
texte = desgaste_funcionario.groupby(['Grau experiência', 'Departamento']).size()
display(texte)
```

```python
#16. Funcionário com mais tempo de empresa e que não esteja demitido
mais_tempo = desgaste_funcionario[
    (desgaste_funcionario['Total de anos trabalhados'] == desgaste_funcionario['Total de anos trabalhados'].max()) &
    (~desgaste_funcionario['Demissão'].str.lower().str.contains('não'))] # ~: Inverta a condição, ou seja, selecione os registros onde "não" não está presente.
display(mais_tempo)
```

```python
#17. Funcionário com menos tempo de empresa e que não esteja demitido
menos_tempo = desgaste_funcionario[
    (desgaste_funcionario['Total de anos trabalhados'] == desgaste_funcionario['Total de anos trabalhados'].min()) &
    (~desgaste_funcionario['Demissão'].str.lower().str.contains('sim'))] # ~: Inverta a condição, ou seja, selecione os registros onde "não" não está presente.
display(menos_tempo)
```
