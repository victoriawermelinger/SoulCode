```python
##### SE O ARQUIVO FOR .CLS####

ip install pandas pyarrow

import pandas as pd

# Carregue o arquivo CSV em um DataFrame
df = pd.read_csv('seu_arquivo.csv')

# Converta o DataFrame para o formato Parquet
df.to_parquet('seu_arquivo.parquet', engine='pyarrow')


##### SE O ARQUIVO FOR .XLSX ####

pip install pandas openpyxl pyarrow

import pandas as pd

# Carregue o arquivo Excel em um DataFrame
df = pd.read_excel('seu_arquivo.xlsx', engine='openpyxl')

# Converta o DataFrame para o formato Parquet
df.to_parquet('seu_arquivo.parquet', engine='pyarrow')



### LEITURA DE ARQUIVOS PARQUET COM SPARK ###

from pyspark.sql import SparkSession

# Inicialize uma SparkSession
spark = SparkSession.builder \
    .appName("Ler Parquet") \
    .getOrCreate()

# Leia o arquivo Parquet
df = spark.read.parquet('seu_arquivo.parquet')

# Exiba as primeiras linhas do DataFrame
df.show()
```
