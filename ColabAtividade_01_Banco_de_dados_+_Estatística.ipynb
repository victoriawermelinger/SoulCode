```python
# Contar o número de aluguéis por cliente
alugueis_por_cliente = aluguel_cliente_df['codcliente'].value_counts()
display(alugueis_por_cliente)

# Calcular a média de aluguéis por cliente
media_alugueis_por_cliente = alugueis_por_cliente.mean()
print(f'Média de aluguéis por cliente: {media_alugueis_por_cliente:.2f}')
```

### **LOCADORA - NOVA IDEIA**

```python
# Outer join entre aluguel e cliente
df_aluguel_cliente = pd.merge(df_aluguel, df_cliente, on='codcliente', how='outer')

# Outer join entre o resultado anterior e a tabela carro
df_aluguel_cliente_carro = pd.merge(df_aluguel_cliente, df_carro, on='codcarro', how='outer')

# Outer join entre o resultado anterior e a tabela marca
df_locadora = pd.merge(df_aluguel_cliente_carro, df_marca, on='codmarca', how='outer')

# Exibir o DataFrame final
display(df_locadora)
```

```python
print(f"Colunas em df_categorias:{df_categorias.columns}")

print(f"\nColunas em df_clientes: {df_clientes.columns}")

print(f"\nColunas em df_fornecedores:{df_fornecedores.columns}")

print(f"\nColunas em df_funcionarios:{df_funcionarios.columns}")

print(f"\nColunas em df_produtos:{df_produtos.columns}")

print(f"\nColunas em df_vendas:{df_vendas.columns}")
```

```python
# Renomeando as colunas problemáticas antes de fazer o merge
df_vendas_renomeado = df_vendas.rename(columns={'id_cliente': 'id_cliente_venda', 'id_funcionario': 'id_funcionario_venda'})
df_produtos_renomeado = df_produtos.rename(columns={'id_categoria': 'id_categoria_produto', 'id_fornecedor': 'id_fornecedor_produto'})

# Fazendo o merge
df_loja = df_vendas_renomeado.merge(df_produtos_renomeado, on='id_produto', how='outer', suffixes=('', '_produto')) \
                              .merge(df_clientes, left_on='id_cliente_venda', right_on='id_cliente', how='outer', suffixes=('', '_cliente')) \
                              .merge(df_funcionarios, left_on='id_funcionario_venda', right_on='id_funcionario', how='outer', suffixes=('', '_funcionario')) \
                              .merge(df_fornecedores, left_on='id_fornecedor_produto', right_on='id_fornecedor', how='outer', suffixes=('', '_fornecedor')) \
                              .merge(df_categorias, left_on='id_categoria_produto', right_on='id_categoria', how='outer', suffixes=('', '_categoria'))

# Exibindo o DataFrame resultante
display(df_loja)
```

```python
# Estatísticas descritivas
display(df_loja.describe())
```

```python
# Contagem de valores únicos em algumas colunas
contagem_unicos = df_loja[['nome', 'cargo', 'nome_categoria', 'nome_fornecedor']].nunique()
print(contagem_unicos)
```

```python
# Custo total de vendas
df_loja['custo_total'] = df_loja['preco'] * df_loja['quantidade']
custo_total = df_loja['custo_total'].sum()
print(f"Custo Total: {custo_total}")
```

```python
# Vendas totais por categoria
vendas_por_categoria = df_loja.groupby('nome_categoria')['quantidade'].sum().sort_values(ascending=False)
print(vendas_por_categoria)

# Visualizando as vendas por categoria
vendas_por_categoria.plot(kind='bar')
plt.title('Vendas Totais por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Total de Vendas')
plt.xticks(rotation=45)
plt.show()
```

```python
# Vendas por funcionário
vendas_por_funcionario = df_loja.groupby('nome')['quantidade'].sum().sort_values(ascending=False)
display(vendas_por_funcionario)

# Visualizando as vendas por Produtos
vendas_por_funcionario.plot(kind='bar')
plt.title('Vendas Totais por Produtos')
plt.xlabel('Produtos')
plt.ylabel('Total de Vendas')
plt.xticks(rotation=90)
plt.show()
```

```python
# Convertendo a coluna de data para datetime, se ainda não estiver
df_loja['data'] = pd.to_datetime(df_loja['data'])

# Vendas por mês
vendas_por_mes = df_loja.groupby(df_loja['data'].dt.to_period('M'))['quantidade'].sum()
print(vendas_por_mes)

# Visualizando vendas por mês
vendas_por_mes.plot(kind='line')
plt.title('Vendas Mensais')
plt.xlabel('Mês')
plt.ylabel('Total de Vendas')
plt.xticks(rotation=45)
plt.show()
```

```python
# Selecionar apenas colunas numéricas
df_numerico = df_loja.select_dtypes(include=['float64', 'int64'])

# Calcular a matriz de correlação
correlation_matrix = df_numerico.corr()
display(correlation_matrix)

# Visualizando a matriz de correlação
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Matriz de Correlação')
plt.show()
```

```python
# Exemplo de boxplot para visualizar a relação entre preço e categoria
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_loja, x='nome_categoria', y='preco')
plt.title('Distribuição de Preços por Categoria')
plt.xticks(rotation=45)
plt.show()
```
