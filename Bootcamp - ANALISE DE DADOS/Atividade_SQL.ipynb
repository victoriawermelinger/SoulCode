### **Desafio 1: Listar Empregados de um Departamento Específico**
1. Liste todos os empregados que pertencem ao departamento cujo nome é
'Marketing'.<p>
2. Crie uma view chamada View_Marketing_Empregados que contenha
todos os empregados do departamento 'Marketing'.

```python
# 1.
SELECT DEPARTAMENTO_ID, nome_empregado, SALARIO, DATA_ADMISSAO, CARGO
FROM EMPREGADO
WHERE DEPARTAMENTO_ID = (
    SELECT DEPARTAMENTO_ID
    FROM DEPARTAMENTO
    WHERE NOME_DEPARTAMENTO = 'Marketing');

```

image.png

```python
# 2.
CREATE VIEW View_Marketing_Empregados AS
SELECT e.EMPREGADO_ID, e.NOME_EMPREGADO, e.SALARIO, e.DATA_ADMISSAO, e.GERENTE_ID, e.CARGO
FROM EMPREGADO e
JOIN DEPARTAMENTO d ON e.DEPARTAMENTO_ID = d.DEPARTAMENTO_ID
WHERE d.Nome_DEPARTAMENTO = 'Marketing';
```

### **Desafio 2: Média de Salário por Departamento**
3. Liste o nome do departamento e a média dos salários de seus
empregados.<p>
4. Crie uma view chamada View_Media_Salarios que mostre o nome do
departamento e a média dos salários dos empregados.

```python
# 3
SELECT d.Nome_DEPARTAMENTO, AVG(e.SALARIO) AS Media_Salario
FROM DEPARTAMENTO d
JOIN EMPREGADO e ON d.DEPARTAMENTO_ID = e.DEPARTAMENTO_ID
GROUP BY d.Nome_DEPARTAMENTO;
```

```python
# 4
CREATE VIEW View_Media_Salarios AS
SELECT d.Nome_DEPARTAMENTO, AVG(e.SALARIO) AS Media_Salario
FROM DEPARTAMENTO d
JOIN EMPREGADO e ON d.DEPARTAMENTO_ID = e.DEPARTAMENTO_ID
GROUP BY d.Nome_DEPARTAMENTO;
```

### **Desafio 3: Empregados com Salário Acima da Média**
5. Liste todos os empregados que têm um salário acima da média dos
salários do seu departamento.
6. Crie uma view chamada View_Empregados_Acima_Media que contenha
os empregados que têm um salário acima da média do seu departamento.

```python
# 5
SELECT e.NOME_EMPREGADO, e.SALARIO, d.Nome_DEPARTAMENTO
FROM EMPREGADO e
JOIN DEPARTAMENTO d ON e.DEPARTAMENTO_ID = d.DEPARTAMENTO_ID
JOIN (
    SELECT DEPARTAMENTO_ID, AVG(SALARIO) AS Media_Salario
    FROM EMPREGADO
    GROUP BY DEPARTAMENTO_ID
) m ON e.DEPARTAMENTO_ID = m.DEPARTAMENTO_ID
WHERE e.SALARIO > m.Media_Salario;
```

image.png

```python
# 6
CREATE VIEW View_Empregados_Acima_Media AS
WITH Media_Salarios AS (
    SELECT DEPARTAMENTO_ID, AVG(SALARIO) AS Media_Salario
    FROM EMPREGADO
    GROUP BY DEPARTAMENTO_ID
)
SELECT e.NOME_EMPREGADO, e.SALARIO, d.Nome_DEPARTAMENTO
FROM EMPREGADO e
JOIN DEPARTAMENTO d ON e.DEPARTAMENTO_ID = d.DEPARTAMENTO_ID
JOIN Media_Salarios m ON e.DEPARTAMENTO_ID = m.DEPARTAMENTO_ID
WHERE e.SALARIO > m.Media_Salario;
```

### **Desafio 4: Departamentos com Mais de X Empregados**
7. Liste todos os departamentos que têm mais de 5 empregados.
8. Crie uma view chamada
View_Departamentos_Com_Muitos_Empregados que contenha os
departamentos com mais de 5 empregados.

```python
# 7
SELECT d.Nome_DEPARTAMENTO, COUNT(e.EMPREGADO_ID) AS Numero_Empregados
FROM DEPARTAMENTO d
JOIN EMPREGADO e ON d.DEPARTAMENTO_ID = e.DEPARTAMENTO_ID
GROUP BY d.Nome_DEPARTAMENTO
HAVING COUNT(e.EMPREGADO_ID) > 5;
```

## **Desafio 10: Empregados de um Departamento com Condição de Salário**
19.Liste todos os empregados do departamento 'Finance' que têm um salário
acima da média dos salários dos empregados desse departamento.<p>
20.Crie uma view chamada View_Finance_Salarios_Acima_Media que
contenha os empregados do departamento 'Finance' com salário acima
da média.

```python
# 19 -- na minha tabela o financeiro e Contabilidade
SELECT e.NOME_EMPREGADO, e.SALARIO
FROM EMPREGADO e
JOIN DEPARTAMENTO d ON e.DEPARTAMENTO_ID = d.DEPARTAMENTO_ID
WHERE d.Nome_DEPARTAMENTO = 'Contabilidade'
AND e.SALARIO > (
    SELECT AVG(e2.SALARIO)
    FROM EMPREGADO e2
    JOIN DEPARTAMENTO d2 ON e2.DEPARTAMENTO_ID = d2.DEPARTAMENTO_ID
    WHERE d2.Nome_DEPARTAMENTO = 'Contabilidade'
);
```

image.png

```python
# 20
CREATE VIEW View_Finance_Salarios_Acima_Media AS
SELECT e.NOME_EMPREGADO, e.SALARIO
FROM EMPREGADO e
JOIN DEPARTAMENTO d ON e.DEPARTAMENTO_ID = d.DEPARTAMENTO_ID
WHERE d.Nome_DEPARTAMENTO = 'Finance'
AND e.SALARIO > (
    SELECT AVG(e2.SALARIO)
    FROM EMPREGADO e2
    JOIN DEPARTAMENTO d2 ON e2.DEPARTAMENTO_ID = d2.DEPARTAMENTO_ID
    WHERE d2.Nome_DEPARTAMENTO = 'Finance'
);
```
