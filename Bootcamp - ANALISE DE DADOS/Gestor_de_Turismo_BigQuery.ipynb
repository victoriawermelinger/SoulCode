## REFERENTE A 2018

```python
import pandas as pd
from io import StringIO
import requests

# Baixar o arquivo
url = 'https://storage.googleapis.com/soulcode_atividade_05/Dados_brutos/cgt-2018.csv'
response = requests.get(url)

# Ler o conteúdo como texto
data = StringIO(response.text)

# Tentar ler com pandas
dfpd_2018 = pd.read_csv(data)
print(dfpd_2018.head())
```

```python


# Renomear as colunas para corrigir os nomes
dfpd_2018.columns = ['Nacionalidade', 'Estado', 'Cidade', 'Bairro', 'Trabalha',
              'Carteira Assinada', 'Tipo de Estabelecimento', 'Sexo',
              'Escolaridade', 'Raça', 'Data de Nascimento', 'Pontuação',
              'Percentual de Conclusão do Canal', 'Data de Conclusão do Canal']
dfpd_2018.columns = [col.capitalize() for col in dfpd_2018.columns]

# Corrigir os caracteres especiais em todo o DataFrame
dfpd_2018.replace({
    'Ã³': 'ó',
    'Ã§': 'ç',
    'Ã£': 'ã',
    'Ãª': 'ê',
    'Ã©': 'é',
    'Ã ': 'à',
    'Ã­': 'í',
    'Ã±': 'ñ',
    'Ãº': 'ú',
    'Ã©': 'é',
    'Ã³': 'ó',
    'Ã ': 'à',
    'Âº': 'º',
    'Âª': 'ª',
    'Â£': '£',
    'Â½': '½',
    'â€“': '–',
    'â€œ': '“',
    'â€': '”',
    'â€': '”',
    'â‚¬': '€',
    'â„¢': '™',
    'â€¢': '•',
    'â€‹': ' ',
}, regex=True, inplace=True)

# Exibir o DataFrame corrigido
display(dfpd_2018)
```

```python
import ftfy

# Aplicando a correção automática de codificação em todas as colunas de string
dfpd_2018 = dfpd_2018.applymap(lambda x: ftfy.fix_text(x) if isinstance(x, str) else x)

# Verificando as mudanças
display(dfpd_2018)
```

```python
import pandas as pd
# Filtrando e deixando apenas o Brasil
dfpd_2018 = dfpd_2018[dfpd_2018['Nacionalidade'] == 'Brasil']
dfpd_2018
```

```python
#ideia para substituir o NaN das duas colunas
# Substituindo valores NaN na coluna 'Trabalha' por 'Não informado'
dfpd_2018['Trabalha'].fillna('Não informado', inplace=True)

# Substituindo valores NaN na coluna 'Carteira Assinada' por 'Não'
dfpd_2018['Carteira assinada'].fillna('Não', inplace=True)
# Verificando as mudanças
display(dfpd_2018)
```
