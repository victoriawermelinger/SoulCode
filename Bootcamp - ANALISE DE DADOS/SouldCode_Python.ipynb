## **1. Vari√°veis**
Vari√°veis em Python (e em muitas outras linguagens de programa√ß√£o) s√£o usadas para armazenar valores que podem ser manipulados e referenciados ao longo do c√≥digo. Pense nelas como "caixas" que guardam informa√ß√µes.

### Conceitos B√°sicos de Vari√°veis

1. **Declara√ß√£o e Atribui√ß√£o**:
   - Em Python, declarar uma vari√°vel e atribuir um valor a ela √© simples. Basta usar o operador de atribui√ß√£o `=`.

   ```python
   x = 5  # Aqui, x √© a vari√°vel e 5 √© o valor atribu√≠do a ela
   nome = "Victoria"  # Vari√°vel nome recebendo uma string
   ```

2. **Tipos de Dados**:
   - Python √© uma linguagem de tipagem din√¢mica, o que significa que voc√™ n√£o precisa declarar o tipo de uma vari√°vel explicitamente. O Python infere o tipo com base no valor atribu√≠do.
   - Exemplos de tipos de dados:
     - **Inteiros**: `int` (e.g., `5`)
     - **Ponto Flutuante**: `float` (e.g., `5.0`)
     - **Strings**: `str` (e.g., `"Victoria"`)
     - **Booleanos**: `bool` (e.g., `True` ou `False`)

   ```python
   idade = 25  # Inteiro
   altura = 1.68  # Ponto flutuante
   cidade = "S√£o Paulo"  # String
   ativo = True  # Booleano
   ```

3. **Manipula√ß√£o de Vari√°veis**:
   - Voc√™ pode realizar opera√ß√µes matem√°ticas e de concatenar strings usando vari√°veis.

   ```python
   a = 10
   b = 20
   soma = a + b  # soma √© 30

   saudacao = "Ol√°, " + nome  # saudacao √© "Ol√°, Victoria"
   ```

4. **Reatribuindo Vari√°veis**:
   - Voc√™ pode mudar o valor de uma vari√°vel ao longo do c√≥digo.

   ```python
   contador = 1
   contador = contador + 1  # Agora, contador √© 2
   ```

### Boas Pr√°ticas com Vari√°veis

- **Nomes Descritivos**: Use nomes de vari√°veis que descrevam seu prop√≥sito.
  
  ```python
  preco_produto = 50.0
  quantidade = 3
  total = preco_produto * quantidade
  ```

- **Evitar Palavras Reservadas**: N√£o use palavras reservadas da linguagem Python como nomes de vari√°veis (e.g., `class`, `if`, `else`, `for`, etc.).

- **Case Sensitivity**: Python diferencia mai√∫sculas de min√∫sculas. Portanto, `idade` e `Idade` seriam vari√°veis diferentes.

## **2. Operadores**

Operadores em Python s√£o s√≠mbolos que realizam opera√ß√µes em vari√°veis e valores. Python possui diversos tipos de operadores, cada um com uma funcionalidade espec√≠fica. Vamos explorar os principais tipos de operadores: aritm√©ticos, de compara√ß√£o, l√≥gicos, de atribui√ß√£o, e outros.

### Operadores Aritm√©ticos
Esses operadores s√£o usados para realizar opera√ß√µes matem√°ticas b√°sicas.

| Operador | Descri√ß√£o                | Exemplo       |
|----------|--------------------------|---------------|
| `+`      | Adi√ß√£o                   | `x + y`       |
| `-`      | Subtra√ß√£o                | `x - y`       |
| `*`      | Multiplica√ß√£o            | `x * y`       |
| `/`      | Divis√£o                  | `x / y`       |
| `%`      | M√≥dulo (resto da divis√£o)| `x % y`       |
| `**`     | Exponencia√ß√£o            | `x ** y`      |
| `//`     | Divis√£o inteira          | `x // y`      |

### Operadores de Compara√ß√£o
Esses operadores comparam dois valores e retornam um valor booleano (`True` ou `False`).

| Operador | Descri√ß√£o                | Exemplo       |
|----------|--------------------------|---------------|
| `==`     | Igual a                  | `x == y`      |
| `!=`     | Diferente de             | `x != y`      |
| `>`      | Maior que                | `x > y`       |
| `<`      | Menor que                | `x < y`       |
| `>=`     | Maior ou igual a         | `x >= y`      |
| `<=`     | Menor ou igual a         | `x <= y`      |

### Operadores L√≥gicos
Esses operadores s√£o usados para combinar express√µes booleanas.

| Operador | Descri√ß√£o                | Exemplo                |
|----------|--------------------------|------------------------|
| `and`    | Retorna `True` se ambas as express√µes forem verdadeiras | `x < 5 and y > 2`  |
| `or`     | Retorna `True` se pelo menos uma das express√µes for verdadeira | `x < 5 or y > 2`  |
| `not`    | Inverte o valor booleano da express√£o | `not(x < 5 and y > 2)`  |

### Operadores de Atribui√ß√£o
Esses operadores s√£o usados para atribuir valores a vari√°veis.

| Operador | Descri√ß√£o                | Exemplo       |
|----------|--------------------------|---------------|
| `=`      | Atribui√ß√£o               | `x = 5`       |
| `+=`     | Atribui√ß√£o com adi√ß√£o    | `x += 3`  (equivale a `x = x + 3`)  |
| `-=`     | Atribui√ß√£o com subtra√ß√£o | `x -= 3`  (equivale a `x = x - 3`)  |
| `*=`     | Atribui√ß√£o com multiplica√ß√£o | `x *= 3`  (equivale a `x = x * 3`)  |
| `/=`     | Atribui√ß√£o com divis√£o   | `x /= 3`  (equivale a `x = x / 3`)  |
| `%=`     | Atribui√ß√£o com m√≥dulo    | `x %= 3`  (equivale a `x = x % 3`)  |
| `//=`    | Atribui√ß√£o com divis√£o inteira | `x //= 3`  (equivale a `x = x // 3`)  |
| `**=`    | Atribui√ß√£o com exponencia√ß√£o | `x **= 3`  (equivale a `x = x ** 3`)  |

### Outros Operadores

#### Operadores de Identidade
Esses operadores comparam a identidade dos objetos, verificando se ocupam a mesma posi√ß√£o na mem√≥ria.

| Operador | Descri√ß√£o                | Exemplo       |
|----------|--------------------------|---------------|
| `is`     | Retorna `True` se ambos os objetos forem o mesmo | `x is y`      |
| `is not` | Retorna `True` se ambos os objetos n√£o forem o mesmo | `x is not y`  |

#### Operadores de Associa√ß√£o
Esses operadores s√£o usados para testar se uma sequ√™ncia cont√©m um determinado valor.

| Operador | Descri√ß√£o                | Exemplo       |
|----------|--------------------------|---------------|
| `in`     | Retorna `True` se o valor estiver na sequ√™ncia | `"a" in "banana"` |
| `not in` | Retorna `True` se o valor n√£o estiver na sequ√™ncia | `"a" not in "banana"` |

## **3. Estruturas Condicionais**

As estruturas condicionais em Python permitem que o programa tome decis√µes baseadas em condi√ß√µes espec√≠ficas, executando diferentes blocos de c√≥digo dependendo dos resultados dessas condi√ß√µes.

### Estrutura `if`

A estrutura `if` avalia uma condi√ß√£o e, se for verdadeira, executa o bloco de c√≥digo correspondente.

```python
idade = 20

if idade >= 18:
    print("Voc√™ √© maior de idade.")
```

### Estrutura `if...else`

A estrutura `if...else` oferece uma alternativa para o caso em que a condi√ß√£o √© falsa.

```python

if idade >= 18:
    print("Voc√™ √© maior de idade.")
else:
    print("Voc√™ √© menor de idade.")
```

### Estrutura `if...elif...else`

A estrutura `if...elif...else` permite testar m√∫ltiplas condi√ß√µes em sequ√™ncia.

```python
nota = 85

    print("Nota: A")
elif nota >= 80:
    print("Nota: B")
elif nota >= 70:
    print("Nota: C")
elif nota >= 60:
    print("Nota: D")
else:
    print("Nota: F")
```

### Estrutura Condicional Aninhada

## **4. Estrutura de Repeti√ß√£o**

As estruturas de repeti√ß√£o em Python permitem que voc√™ execute um bloco de c√≥digo v√°rias vezes. Existem duas principais estruturas de repeti√ß√£o: `for` e `while`.

### **4.1 La√ßo `for`**

O la√ßo `for` √© usado para iterar sobre uma sequ√™ncia (como uma lista, tupla, dicion√°rio, conjunto ou string) ou outros objetos iter√°veis.

#### **Sintaxe B√°sica**

```python
for item in sequencia:
    # bloco de c√≥digo
```

#### **Exemplo com Listas**

```python
numeros = [1, 2, 3, 4, 5]

for numero in numeros:
    print(numero)
```

#### **Exemplo com a Fun√ß√£o `range()`**

A fun√ß√£o `range()` √© frequentemente usada com o la√ßo `for` para gerar uma sequ√™ncia de n√∫meros.

```python
for i in range(5):  # Gera os n√∫meros 0 a 4
    print(i)
```

#### **Exemplo com Strings**

```python
palavra = "Python"

for letra in palavra:
    print(letra)
```

### **4.2 La√ßo `while`**

O la√ßo `while` repete um bloco de c√≥digo enquanto uma condi√ß√£o for verdadeira.

#### **Sintaxe B√°sica**

```python
while condicao:
    # bloco de c√≥digo
```

#### **Exemplo de Contagem Regressiva**

```python
contador = 5

while contador > 0:
    print(contador)
    contador -= 1
```

#### **La√ßo Infinito**

Um la√ßo `while` sem uma condi√ß√£o de t√©rmino pode se tornar um la√ßo infinito.

```python
while True:
    print("Este √© um la√ßo infinito.")
    break  # Use `break` para sair do la√ßo
```

### **4.3 Controle de Fluxo no La√ßo**

#### **Comando `break`**

O comando `break` encerra o la√ßo prematuramente.

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

#### **Comando `continue`**

O comando `continue` pula para a pr√≥xima itera√ß√£o do la√ßo.

```python
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```
