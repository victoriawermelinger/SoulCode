##População e Amostra

*   **População:** refere-se ao conjunto de dados COMPLETO de todos os elementos ou eventos que estamos interessados em estudar.
*   **Amostra:** refere-se a um subconjunto da população, usada para fazer inferências sobre a população.


**Exemplo:**

***População:*** Todos os residentes de uma cidade.

***Amostra:*** 500 residentes selecionados para uma pesquisa de opinião.



```python
import pandas as pd
import numpy as np

# Criando um conjunto de dados com 1000 clientes (população)
clientes = pd.DataFrame({
    'ID': range(1,10001),
    'Gastos': np.round(np.random.normal(500, 100, 10000),2)  # Gastos em uma média de 500 com desvio padrão de 100
})

# display(clientes)

# Extraindo uma amostra aleatória de 100 clientes
amostra = clientes.sample(n=100, random_state=1)
display(amostra)
```

##Média, Mediana e Moda

* **Média:** A soma de todos os valores dividida pelo número de valores. É usada para encontrar o valor "médio" de um conjunto de dados.
* **Mediana:** O valor central em um conjunto de dados ordenado. Ela é útil quando há outliers (valores extremos) que distorcem a média.
* **Moda:** O valor que aparece com maior frequência em um conjunto de dados.


**Exemplo:** Suponha que você tenha os seguintes dados de vendas diárias de uma loja (em unidades): 5, 7, 7, 8, 9.

***Média:*** (5 + 7 + 7 + 8 + 9) / 5 = 7,2

***Mediana:*** 7 (o valor central)

***Moda:*** 7 (aparece mais frequentemente)

```python
# Média
media = amostra['Gastos'].mean()

# Mediana
mediana = amostra['Gastos'].median()

# Moda
moda = amostra['Gastos'].mode()[0]

print(f"Média: {media}, Mediana: {mediana}, Moda: {moda}")
```

##Desvio Padrão e Variância

* **Variância:** Mede o quanto os dados se dispersam em relação à média. É a média dos quadrados das diferenças entre os valores e a média.
* **Desvio Padrão:** A raiz quadrada da variância. Ele indica o quão espalhados estão os dados em torno da média.

A Distribuição Normal (ou Gaussiana) é uma das distribuições mais importantes em estatística. Ela é simétrica, com a maioria dos valores próximos à média, e tem a forma de um sino. Muitos fenômenos naturais seguem essa distribuição, como altura, peso e desempenho.

##Correlação

A Correlação mede a força e a direção de uma relação linear entre duas variáveis. O coeficiente de correlação varia de -1 a 1:

1 indica uma correlação positiva perfeita (à medida que uma variável aumenta, a outra também aumenta).

-1 indica uma correlação negativa perfeita (à medida que uma variável aumenta, a outra diminui).

0 indica que não há correlação linear.

##Intervalo de Confiança

Um intervalo de confiança fornece uma faixa de valores dentro da qual esperamos que um parâmetro da população (como a média) esteja, com uma certa probabilidade.

```python
from scipy import stats

# Definindo o nível de confiança (95%)
nivel_confianca = 0.95
n = len(amostra['Gastos'])
media = amostra['Gastos'].mean()
desvio_padrao = amostra['Gastos'].std()

# Calculando o intervalo de confiança
intervalo_confianca = stats.norm.interval(nivel_confianca, loc=media, scale=desvio_padrao/np.sqrt(n))

print(f"Intervalo de Confiança: {intervalo_confianca}")
```

```python
import pandas as pd
import numpy as np
from scipy import stats

# Criando o DataFrame com números de 1 a 10
numeros = pd.DataFrame({
    'Valores': range(1, 11)  # Valores de 1 a 10
})

# Definindo o nível de confiança (95%)
nivel_confianca = 0.95

# Número de observações (n)
n = len(numeros['Valores'])

# Calculando a média da amostra
media = numeros['Valores'].mean()

# Calculando o desvio padrão da amostra
desvio_padrao = numeros['Valores'].std()

# Calculando o intervalo de confiança
intervalo_confianca = stats.norm.interval(nivel_confianca, loc=media, scale=desvio_padrao/np.sqrt(n))

# Exibindo o intervalo de confiança
print(f"Intervalo de Confiança: {intervalo_confianca}")
```

##Testes de Hipóteses


O teste de hipóteses é usado para avaliar se uma suposição sobre um parâmetro da população é verdadeira. Envolve formular duas hipóteses:

Hipótese nula (H0): uma declaração de que não há efeito ou diferença.

Hipótese alternativa (H1): uma declaração de que há um efeito ou diferença.


```python
# Realizando um teste t de uma amostra
t_stat, p_valor = stats.ttest_1samp(amostra['Gastos'], 500)

print(f"Estatística t: {t_stat}, p-valor: {p_valor}")

# Interpretação do p-valor
if p_valor < 0.05:
    print("Rejeitamos a hipótese nula (H0), a média dos gastos é significativamente diferente de 500.")
else:
    print("Não podemos rejeitar a hipótese nula (H0), a média dos gastos não é significativamente diferente de 500.")
```

```python
import pandas as pd
import numpy as np
from scipy import stats

# Simulando um conjunto de dados de vendas após uma campanha publicitária
np.random.seed(42)  # Para reprodutibilidade
vendas = np.random.normal(10500, 2000, 100)  # Média de R$ 10.500 com desvio padrão de R$ 2000
data = pd.DataFrame({'Vendas': vendas})

# Realizando um teste t de uma amostra
valor_hipotetico = 50000  # Valor hipotético para a média das vendas
t_stat, p_valor = stats.ttest_1samp(data['Vendas'], valor_hipotetico)

# Resultados do teste
print(f"Estatística t: {t_stat}, p-valor: {p_valor}")

# Interpretação do p-valor
if p_valor < 0.05:
    print("Rejeitamos a hipótese nula (H0), a média das vendas é significativamente diferente de R$ 10.000.")
else:
    print("Não podemos rejeitar a hipótese nula (H0), a média das vendas não é significativamente diferente de R$ 10.000.")
```

```python
# Instalar bibliotecas necessárias (caso ainda não tenha)
!pip install pandas numpy scipy openpyxl

import pandas as pd
import numpy as np
import scipy.stats as stats
from google.colab import files

# Upload do arquivo de vendas
uploaded = files.upload()

# Carregar a planilha
vendas_df = pd.read_excel(list(uploaded.keys())[0])

# Exibir as primeiras linhas para verificar o conteúdo
vendas_df.head()

# 1. Média, Mediana e Moda
media = vendas_df['Valor Vendido'].mean()
mediana = vendas_df['Valor Vendido'].median()
moda = vendas_df['Valor Vendido'].mode()[0]  # Moda pode retornar múltiplos valores, pegamos o primeiro

# 2. Desvio Padrão e Variância
desvio_padrao = vendas_df['Valor Vendido'].std()
variancia = vendas_df['Valor Vendido'].var()

# 3. Teste de Normalidade (Shapiro-Wilk)
shapiro_test = stats.shapiro(vendas_df['Valor Vendido'])

# 4. Correlação entre Valor Vendido e Dias
vendas_df['Dias'] = (vendas_df['Data'] - vendas_df['Data'].min()).dt.days
correlacao = vendas_df[['Dias', 'Valor Vendido']].corr().iloc[0, 1]  # Correlação entre 'Dias' e 'Valor Vendido'

# 5. Intervalo de Confiança para a média
conf_interval = stats.t.interval(0.95, len(vendas_df['Valor Vendido'])-1, loc=media, scale=stats.sem(vendas_df['Valor Vendido']))

# 6. Teste de Hipóteses - Média Loja Física vs Internet
grupo_fisica = vendas_df[vendas_df['Plataforma'] == 'Loja Física']['Valor Vendido']
grupo_internet = vendas_df[vendas_df['Plataforma'] == 'Internet']['Valor Vendido']
t_stat, p_value = stats.ttest_ind(grupo_fisica, grupo_internet, equal_var=False)

# Exibir os resultados
print(f"Média: {media}")
print(f"Mediana: {mediana}")
print(f"Moda: {moda}")
print(f"Desvio Padrão: {desvio_padrao}")
print(f"Variância: {variancia}")
print(f"Teste de Normalidade (Shapiro-Wilk): {shapiro_test}")
print(f"Correlação (Dias x Valor Vendido): {correlacao}")
print(f"Intervalo de Confiança (95%) para a Média: {conf_interval}")
print(f"Teste de Hipóteses (Loja Física vs Internet): t_stat={t_stat}, p_value={p_value}")
```

##Teste Qui-Quadrado (Chi-Square)

```python
from scipy import stats

# Criando uma tabela de contingência
tabela = pd.crosstab(data['Estrato'], pd.cut(data['Gastos'], bins=[0, 400, 600, 800, 1000], labels=['Baixo', 'Médio', 'Alto', 'Muito Alto']))
print("Tabela de Contingência:")
print(tabela)

# Realizando o teste qui-quadrado
chi2, p_valor, _, _ = stats.chi2_contingency(tabela)
print(f"Estatística qui-quadrado: {chi2}, p-valor: {p_valor}")
```

##Distribuição Binomial

```python
# Importar bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parâmetros para a distribuição binomial
n = 10  # número de experimentos (exemplo: 10 visitas)
p = 0.5  # probabilidade de sucesso em cada experimento (exemplo: 10% de chance de venda por visita)

# Gerar valores binomiais
x = np.arange(0, n+1)
binom_pmf = binom.pmf(x, n, p)

# Plotar a distribuição binomial
plt.figure(figsize=(8,6))
plt.bar(x, binom_pmf, width=0.1, color='b', alpha=0.7)
plt.xlabel('Número de vendas')
plt.ylabel('Probabilidade')
plt.title(f'Distribuição Binomial (n={n}, p={p})')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Carregar a planilha de exemplo
file_path = '/content/MinhasVendas.xlsx'
df = pd.read_excel(file_path)

# Visualizar as primeiras linhas do dataframe
print("Primeiras linhas do DataFrame:")
print(df.head())

# Calcular a probabilidade média de venda por visita
p = df['Quantidade_de_Vendas'].sum() / df['Numero_de_Visitas'].sum()
print(f"\nProbabilidade média de venda por visita: {p:.2f}")

# Definir o número de visitas (n)
n = 100  # Exemplo: 100 visitas

# Gerar valores binomiais
x = np.arange(0, n+1)
binom_pmf = binom.pmf(x, n, p)

# Plotar a distribuição binomial
plt.figure(figsize=(10,6))
plt.bar(x, binom_pmf, width=0.4, color='b', alpha=0.7)
plt.xlabel('Número de vendas')
plt.ylabel('Probabilidade')
plt.title(f'Distribuição Binomial - {n} visitas, p={p:.2f}')
plt.grid(True)
plt.show()
```

```python

!pip install pandas openpyxl scikit-learn scipy matplotlib seaborn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Carregar a planilha de exemplo
file_path = '/content/MinhasVendas.xlsx'
df = pd.read_excel(file_path)

# Visualizar as primeiras linhas do dataframe
print("Primeiras linhas do DataFrame:")
print(df.head())

# Calcular a probabilidade média de venda por visita
p = df['Quantidade_de_Vendas'].sum() / df['Numero_de_Visitas'].sum()
print(f"\nProbabilidade média de venda por visita: {p:.2f}")

# Definir o número de visitas (n)
n = 100  # Exemplo: 100 visitas

# Gerar valores binomiais
x = np.arange(0, n+1)
binom_pmf = binom.pmf(x, n, p)

# Plotar a distribuição binomial com melhorias
plt.figure(figsize=(14, 8))
bars = plt.bar(x, binom_pmf, width=0.6, color=plt.cm.viridis(binom_pmf / max(binom_pmf)), edgecolor='black')

# Adicionar rótulos e título
plt.xlabel('Número de Vendas', fontsize=14)
plt.ylabel('Probabilidade', fontsize=14)
plt.title(f'Distribuição Binomial - {n} Visitas, p={p:.2f}', fontsize=16)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adicionar anotações nas barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2 - 0.3, yval + 0.001, f'{yval:.3f}', fontsize=9, va='bottom')

# Suavizar a visualização dos dados
window_size = 10  # Tamanho da janela para suavização
window = np.ones(window_size) / window_size
binom_pmf_smooth = np.convolve(binom_pmf, window, mode='same')

# Plotar a linha suavizada
plt.plot(x, binom_pmf_smooth, color='red', linewidth=4)

plt.show()
```
