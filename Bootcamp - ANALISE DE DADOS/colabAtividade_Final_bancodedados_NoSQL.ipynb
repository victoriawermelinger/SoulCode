```python
# Função para adicionar um novo livro
def adicionar_livro():
    titulo = input("Digite o título do livro: ")
    autor = input("Digite o autor do livro: ")
    ano = int(input("Digite o ano de publicação: "))
    isbn = input("Digite o ISBN do livro: ")

    livro = {
        "titulo": titulo,
        "autor": autor,
        "ano": ano,
        "isbn": isbn
    }

    collection.insert_one(livro)
    print("Livro adicionado com sucesso!")

# Função para listar todos os livros cadastrados
def listar_livros():
    livros = collection.find()
    for livro in livros:
        print(f"Título: {livro['titulo']}, Autor: {livro['autor']}, Ano: {livro['ano']}, ISBN: {livro['isbn']}")

# Função para buscar livro por título, autor ou ISBN
def buscar_livro():
    criterio = input("Você quer buscar por (1) título, (2) autor ou (3) ISBN? ")

    if criterio == '1':
        titulo = input("Digite o título do livro: ")
        livros = collection.find({"titulo": titulo})
    elif criterio == '2':
        autor = input("Digite o autor do livro: ")
        livros = collection.find({"autor": autor})
    elif criterio == '3':
        isbn = input("Digite o ISBN do livro: ")
        livros = collection.find({"isbn": isbn})
    else:
        print("Opção inválida.")
        return

    for livro in livros:
        print(f"Título: {livro['titulo']}, Autor: {livro['autor']}, Ano: {livro['ano']}, ISBN: {livro['isbn']}")

# Função para atualizar informações de um livro
def atualizar_livro():
    isbn = input("Digite o ISBN do livro que deseja atualizar: ")
    livro = collection.find_one({"isbn": isbn})

    if livro:
        novo_titulo = input(f"Novo título (atual: {livro['titulo']}): ") or livro['titulo']
        novo_autor = input(f"Novo autor (atual: {livro['autor']}): ") or livro['autor']
        novo_ano = input(f"Novo ano (atual: {livro['ano']}): ") or livro['ano']

        collection.update_one(
            {"isbn": isbn},
            {"$set": {"titulo": novo_titulo, "autor": novo_autor, "ano": novo_ano}}
        )
        print("Livro atualizado com sucesso!")
    else:
        print("Livro não encontrado.")

# Função para remover um livro do banco de dados
def remover_livro():
    isbn = input("Digite o ISBN do livro que deseja remover: ")
    resultado = collection.delete_one({"isbn": isbn})

    if resultado.deleted_count > 0:
        print("Livro removido com sucesso!")
    else:
        print("Livro não encontrado.")

# Menu principal
def menu():
    while True:
        print("\nSistema de Gerenciamento de Livros")
        print("1. Adicionar Livro")
        print("2. Listar Livros")
        print("3. Buscar Livro")
        print("4. Atualizar Livro")
        print("5. Remover Livro")
        print("6. Sair")

        opcao = input("Escolha uma opção: ")

        if opcao == '1':
            adicionar_livro()
        elif opcao == '2':
            listar_livros()
        elif opcao == '3':
            buscar_livro()
        elif opcao == '4':
            atualizar_livro()
        elif opcao == '5':
            remover_livro()
        elif opcao == '6':
            print("Saindo do sistema...")
            break
        else:
            print("Opção inválida, tente novamente.")

# Executar o sistema
if __name__ == "__main__":
    menu()
```
