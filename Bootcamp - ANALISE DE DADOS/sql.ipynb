```python
-- FUNÇÕES DE AGREGAÇÃO
-- AVG (MEDIA), MIN, MAX, COUNT, SUM

-- MEDIA SALARIO
SELECT AVG(SALARIO) AS MEDIA_SALARIAL FROM EMPREGADO

-- MAIOR SALARIO
SELECT MAX(SALARIO) AS MAIOR_SALARIO FROM EMPREGADO

-- MENOR SALARIO
SELECT MIN(SALARIO) AS MENOR_SALARIO FROM EMPREGADO

-- CONTAR O TOTAL DE SALARIOS
SELECT COUNT(SALARIO) AS TOTAL_SALARIO FROM EMPREGADO

--SOMA TOTAL DE SALARIOS
SELECT SUM(SALARIO) AS SOMA_TOTAL_SALARIO FROM EMPREGADO




-- COMBINAR AS FUNÇÕES DE AGREGAÇÃO
SELECT
  AVG(SALARIO) AS MEDIA_SALARIAL,
  MAX(SALARIO) AS MAIOR_SALARIO,
  MIN(SALARIO) AS MENOR_SALARIO,
  COUNT(SALARIO) AS TOTAL_SALARIO,
  SUM(SALARIO) AS SOMA_TOTAL_SALARIO
FROM EMPREGADO


-- USANDO SUBCONSULTAS (ALTERNATIVA)
SELECT
    (SELECT AVG(SALARIO) FROM EMPREGADO) AS MEDIA_SALARIAL,
    (SELECT MAX(SALARIO) FROM EMPREGADO) AS MAIOR_SALARIO,
    (SELECT MIN(SALARIO) FROM EMPREGADO) AS MENOR_SALARIO,
    (SELECT COUNT(SALARIO) FROM EMPREGADO) AS TOTAL_SALARIO,
    (SELECT SUM(SALARIO) FROM EMPREGADO) AS SOMA_TOTAL_SALARIO;


-- CROSS JOIN
SELECT
    Metrics.MEDIA_SALARIAL,
    Metrics.MAIOR_SALARIO,
    Metrics.MENOR_SALARIO,
    Metrics.TOTAL_SALARIO,
    Metrics.SOMA_TOTAL_SALARIO
FROM
    (SELECT
        AVG(SALARIO) AS MEDIA_SALARIAL,
        MAX(SALARIO) AS MAIOR_SALARIO,
        MIN(SALARIO) AS MENOR_SALARIO,
        COUNT(SALARIO) AS TOTAL_SALARIO,
        SUM(SALARIO) AS SOMA_TOTAL_SALARIO
    FROM EMPREGADO) AS Metrics;

-- UNION ALL
SELECT 'MEDIA_SALARIAL' AS METRICA, AVG(SALARIO) AS VALOR FROM EMPREGADO
UNION ALL
SELECT 'MAIOR_SALARIO', MAX(SALARIO) FROM EMPREGADO
UNION ALL
SELECT 'MENOR_SALARIO', MIN(SALARIO) FROM EMPREGADO
UNION ALL
SELECT 'TOTAL_SALARIO', COUNT(SALARIO) FROM EMPREGADO
UNION ALL
SELECT 'SOMA_TOTAL_SALARIO', SUM(SALARIO) FROM EMPREGADO;



```

```python
#BANCO DE DADOS EMPRESADB

CREATE TABLE IF NOT EXISTS DEPARTAMENTO (
    numdepto INT AUTO_INCREMENT PRIMARY KEY,
    nome VARCHAR(50) NOT NULL
);

CREATE TABLE IF NOT EXISTS EMPREGADO (
    matricula INT AUTO_INCREMENT PRIMARY KEY,
    nome VARCHAR(70) NOT NULL,
    salario DECIMAL(10,2),
    numdepto INT,
    CONSTRAINT EMPREGADO_FK FOREIGN KEY (numdepto) REFERENCES DEPARTAMENTO(numdepto)
);

CREATE TABLE IF NOT EXISTS PROJETO (
    codprojeto INT AUTO_INCREMENT PRIMARY KEY,
    nome VARCHAR(100) NOT NULL,
    data_inicio DATE,
    data_fim DATE
);

CREATE TABLE IF NOT EXISTS TRABALHA_EM (
    matricula INT,
    codprojeto INT,
    data_inicio DATE,
    data_fim DATE,
    PRIMARY KEY (matricula, codprojeto),
    CONSTRAINT TRABALHA_FK FOREIGN KEY (matricula) REFERENCES EMPREGADO(matricula),
    CONSTRAINT TRABALHA_FK1 FOREIGN KEY (codprojeto) REFERENCES PROJETO(codprojeto)
);

CREATE TABLE IF NOT EXISTS HABILIDADE (
    codhabilidade INT AUTO_INCREMENT PRIMARY KEY,
    descricao VARCHAR(100) NOT NULL
);

CREATE TABLE IF NOT EXISTS EMPREGADO_HABILIDADE (
    matricula INT,
    codhabilidade INT,
    PRIMARY KEY (matricula, codhabilidade),
    CONSTRAINT EMPREGADO_HAB_FK FOREIGN KEY (matricula) REFERENCES EMPREGADO(matricula),
    CONSTRAINT EMPREGADO_HAB_FK1 FOREIGN KEY (codhabilidade) REFERENCES HABILIDADE(codhabilidade)
);

```

#Operações JOIN


- Combinar linhas de duas ou mais tabelas, baseadas em condição relacionada entre elas.

##Tipos de JOINs


*   **INNER JOIN:** Retorna apenas as linhas que têm correspondências em ambas as tabelas.
*   **LEFT JOIN (ou LEFT OUTER JOIN):** Retorna todas as linhas da tabela à esquerda e as linhas correspondentes da tabela à direita. Se não houver correspondência, os resultados conterão NULL para as colunas da tabela à direita.
*  **RIGHT JOIN (ou RIGHT OUTER JOIN):** Retorna todas as linhas da tabela à direita e as linhas correspondentes da tabela à esquerda. Se não houver correspondência, os resultados conterão NULL para as colunas da tabela à esquerda.
*   **CROSS JOIN:** Retorna o produto cartesiano das linhas das tabelas envolvidas, ou seja, todas as combinações possíveis de linhas.
*   **FULL JOIN (ou FULL OUTER JOIN):** Retorna todas as linhas quando há uma correspondência em uma das tabelas.
