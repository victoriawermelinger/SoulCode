## Começo dos Exercícios: 

```python
#1. Contando a frequência de viagens de negócios
data_f['quantidade_viagem'] = data_f['Viagem de negócios'].value_counts()

display(quantidade_viagem)

#desgaste_funcionario.to_csv('atividade1', index =False)
```

```python
#2. Área de formação dos Funcionarios 
area_func = data_f['Área de Formação'].value_counts()
display(area_func)

#desgaste_funcionario.to_csv('atividade2', index =False)
#desgaste_funcionario.to_csv(os.path.join(r'C:\Users\victoria\Documents\MeusProjetos- GIT\SoulCode\Bootcamp - ANALISE DE DADOS\Atividades_Pandas03', 'atividade2.csv'), index=False)
```

```python
#3. Agrupando por 'Área de Formação' e 'Viagem de negócios' e contando o número de ocorrências

contagem_area_viagem = data_f.groupby(['Área de Formação','Viagem de negócios']).size().unstack()
display(contagem_area_viagem)
```

```python
#4. Departamento que tem mais viagens 
contagem_area_viagem = data_f.groupby(['Departamento','Viagem de negócios']).size().unstack()
display(contagem_area_viagem)
```

```python
#5. Quadro de formação por departamento 
contagem_dep_area = data_f.groupby(['Área de Formação', 'Departamento']).size().unstack()
display(contagem_dep_area)
```

```python
#6. Remuneração por ações da empresa para os funcionarios
def remuneração(row):
        if row['opção de remaunração variável em ações'] == 3:
            return 'Participação nas ações'
        elif row['opção de remaunração variável em ações'] == 2:
            return 'Recebe dividendos, mas não tem ação'
        elif row['opção de remaunração variável em ações'] == 1:
            return 'Recebe dividendos'
        else:
            return 'Não recebe'

data_f['Grau Remuneração'] = data_f.apply(remuneração, axis=1)
display(data_f[['Departamento', 'Área de Formação', 'Grau Remuneração']])
```

```python
#7. contagem da remuneração dos funcionários 
grau_remuneração = data_f[('Grau Remuneração')].value_counts()
display(grau_remuneração)
```

```python
#8. grau de satisfação dos funcionarios (Nível de Satisfação com o trabalho)
def Satisfação(row):
        if row['Satisfação com o trabalho'] == 4:
            return 'Muito Alta'
        elif row['Satisfação com o trabalho'] == 3:
            return 'Alta'
        elif row['Satisfação com o trabalho'] == 2:
            return 'Média'
        else:
            return 'Baixa'

data_f['Grau Satisfação'] = data_f.apply(Satisfação, axis=1)
display(data_f[['Departamento', 'Área de Formação', 'Grau Satisfação']])
```

```python
#9. contagem de Grau Satisfação dos funcionários 
grau_satisfação = data_f[('Grau Satisfação')].value_counts()
display(grau_satisfação)
```

```python
#10.  grau de escolaridade dos funcionarios 
def Escolaridade(row):
        if row['Escolaridade'] == 5:
            return 'Doutor'
        elif row['Escolaridade'] == 4:
            return 'Mestre'
        elif row['Escolaridade'] == 3:
            return 'Bacharelado'
        elif row['Escolaridade'] == 2:
            return 'Faculdade'
        else:
            return 'Abaixo da faculdade'

data_f['Grau Escolaridade'] = data_f.apply(Escolaridade, axis=1)

display(data_f[['Área de Formação', 'Grau Escolaridade']])
```

```python
#11. contagem de Grau escolaridade dos funcionários 
grau_escolaridade_func = data_f['Grau Escolaridade'].value_counts()
display(grau_escolaridade_func)
```

```python
#12. grau de Experiência dos funcionarios 

def experiência(row):
    if row['Total de anos trabalhados'] > 0 and row['Total de anos trabalhados'] <= 5:
        return 'JV/Estágiario'
    elif row['Total de anos trabalhados'] > 5 and row['Total de anos trabalhados'] <= 10:
        return 'Junior'
    elif row['Total de anos trabalhados'] > 10 and row['Total de anos trabalhados'] <= 20:
        return 'Pleno'
    elif row['Total de anos trabalhados'] > 20:
        return 'Senior'
    else:
        return 'Sem experiência'

data_f['Grau experiência'] = data_f.apply(experiência, axis=1)

display(data_f[['Total de anos trabalhados', 'Área de Formação', 'Grau experiência']])
```

```python
#13. Contagem dos diferentes valores na coluna 'Grau experiência'
grau_experiência_func = data_f['Grau experiência'].value_counts()
display(grau_experiência_func)
```

```python
#14. Calcular a média dos anos trabalhados para cada grau de experiência
media_exp = data_f.groupby('Grau experiência')['Total de anos trabalhados'].mean()
display(media_exp)
```

```python
#15. Contar o número de funcionários em cada combinação de 'Grau experiência' e 'Departamento'
grau_exp_dp = data_f.groupby(['Grau experiência', 'Departamento']).size()
display(grau_exp_dp)
```

```python
#16. Funcionário com mais tempo de empresa e que não esteja demitido
# ~: Inverta a condição, ou seja, selecione os registros onde "não" não está presente.
#.str.contains('sim'): verifica se a string contém a palavra "não"
#.str.lower(): converte a string para letras minúsculas

mais_tempo = data_f[
    (data_f['Total de anos trabalhados'] == data_f['Total de anos trabalhados'].max()) &
    (~data_f['Demissão'].str.lower().str.contains('não'))] 
display(mais_tempo)
```

```python
#17. Funcionário com menos tempo de empresa e que não esteja demitido
# ~: Inverta a condição, ou seja, selecione os registros onde "não" não está presente.
#.str.contains('sim'): verifica se a string contém a palavra "sim"
#.str.lower(): converte a string para letras minúsculas

menos_tempo = data_f[
    (data_f['Total de anos trabalhados'] == data_f['Total de anos trabalhados'].min()) &
    (~data_f['Demissão'].str.lower().str.contains('sim'))] 
display(menos_tempo)
```

```python
#18. Calcular a média do salário mensal por gênero
Salario_genero = data_f.groupby('Gênero')['Salário Mensal'].mean()

# Adiciona a média do salário por gênero ao DataFrame original
data_f['Salario_genero'] = data_f['Gênero'].map(Salario_genero)

# Exibe as colunas relevantes
display(data_f[['Departamento','Área de Formação' ,'Gênero', 'Salário Mensal', 'Salario_genero']])
```

```python
# Calcula a média do salário por departamento e atribui a cada linha
data_f['méd_salario_Dp'] = data_f.groupby('Departamento')['Salário Mensal'].mean()

display(data_f[['Departamento','Área de Formação', 'Salário Mensal', 'méd_salario_Dp']])
```

```python
#20 Aplicar Funções Lambda a Multiplas colunas
data_f['soma_salario_idade'] = data_f.apply(lambda linha: linha['Salário Mensal'] + linha['Idade'], axis=1)
display(data_f[['Departamento','soma_salario_idade']])
```

```python
# 21. Imposto pago em cima das diárias recebidas
def calcular_imposto(salario):
    return salario * 0.01

data_f['Imposto'] = data_f['Salário Mensal'].apply(calcular_imposto)
colunas_a_mostrar = data_f[['Salário Mensal', 'Imposto']]
display(colunas_a_mostrar)
```

```python
#22. Custo de total de viagens por departamento 
custo_viagem_DP = data_f.groupby('Departamento')['Diárias recebidas'].sum()
display(custo_viagem_DP)
```

```python
#23. Funcionário mais velho e qual é o departamento e quanto tempo de com o gestor atual.
func_mais_antigo = data_f[['Idade','Departamento', 'Total de anos trabalhados', 'ID Funcionário', 'Área de Formação']].max()
display(func_mais_antigo)
```

```python
#24. Funcionário mais novo e qual é o departamento e quanto tempo de com o gestor atual.
func_mais_novo = data_f[['Idade','Departamento', 'Total de anos trabalhados', 'ID Funcionário', 'Área de Formação']].min()
display(func_mais_novo)
```

```python
#25.  Supondo que data_f seja o seu DataFrame original
df_filtrar_demissao1 = data_f[['Demissão', 'Gênero']]

# Agrupando por 'Gênero' e 'Demissão' e contando as ocorrências
contagem_demisao_genero = df_filtrar_demissao1.groupby(['Gênero', 'Demissão']).size().reset_index(name='Contagem')
display(contagem_demisao_genero)
```

```python
#26. Calcular o total de pessoas por gênero
total_genero = df_filtrar_demissao1.groupby('Gênero').size().reset_index(name='Total')
print('Total de funcionários por gênero e relalção de demissões')
display(total_genero)

# Unir os dois DataFrames com base no gênero
df_merged = pd.merge(contagem_demisao_genero, total_genero, on='Gênero')
display(df_merged)

# Calcular a porcentagem de demissões
df_merged['Porcentagem'] = (df_merged['Contagem'] / df_merged['Total']) * 100
display(df_merged)

# Filtrar para mostrar apenas as demissões
df_porcentagem_demissao = df_merged[df_merged['Demissão'] == 'Sim']
display(df_porcentagem_demissao)

for index, row in df_porcentagem_demissao.iterrows():
    genero = row['Gênero']
    porcentagem = row['Porcentagem']
    print(f"A porcentagem de demissões entre {genero}s é de {porcentagem:.2f}%.")
```
